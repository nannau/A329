# Web-Exploration of Numerical Data Interactive - WENDI

## Exploring stellar simulations
This platform allows users to explore and analyse different types of simulation data sets. The following areas are available:

### [NuGrid](http://www.nugridstars.org) stellar nucleosynthesis data
This area is still at the [original NuGrid Wendi](http://wendi.nugridstars.org) and will be moved over to this new platform soon.
* [NuGridPy](https://nugrid.github.io/NuGridPy) access to the NuGrid stellar evolution, explosion and nucleosynthesis data
* [NuPyCee](http://nugrid.github.io/NuPyCEE)

### PPMstar stellar hydro data from [LCSE](http://www.lcse.umn.edu) and [CSA](http://csa.phys.uvic.ca/research/stellar-hydrodynamics)
This area has at this point a few test and demo notebooks. It is planned to make more data sets and analysis tools available.

* [PPMstar area](./PPMstar)

### Available software

ubuntu, unzip, vim, wget

#### jupyter notebooks
astropy, keras, matplotlib, mpmath, numpy, python 2 and 3 kernels,  seaborn, setuptools, scikit-learn, scikit-image, scipy, sympy, tensorflow, folium 

### Terminal
When using a _Terminal_ (see _New_ dropdown menu on top right) presently the loing shell is `sh`. In order to switch to the much nicer _bash_ shell, just type `bash` at the command line. 

### What is available in the jupyter home section?
When you are on the home screen of the jupyter session you will see the content that is located in your home directory's `notebooks` directory. 

### Platform updates
When new capabilties or updates are required in the jupyter sessions the platform administrator will build a new jupyter session image. Your currently running instance of the jupyter session image is not effected and will not have these updates until you restart your session. In order to do this proceed as follows:

1. save your home directory to one of the persistent storage options using the terminal and appropriate terminal commands (e.g. `git`, `rsync` or `ssh`)
2. go to the home screen of the jupyter session and click `Logout` in the top right corner
3. restart the instance by first clicking on the red `Stop My Server` button to stop the instance; the server needs about 15s to shut down and it is best to wait that moment, and then ...
4. on the next screen select the green `My Server` button (you may get an _Error: 500 ..._ just reload the wendi URL and it should work)
5. your session should now run the latest docker image
6. recover your files in your home directory using the terminal


## About
This is the Web-Exploration of Numerical Data Interactive (WENDI)
platform developed and maintained by the [Computational Stellar
Astrophysics (CSA)](http://csa.phys.uvic.ca) group in the [Department
of Physics and Astronomy](http://www.uvic.ca/science/physics) at the
[University of Victoria (UVic)](http://www.uvic.ca). CSA is part of
the UVic [Astronomy Research Center
(ARC)](http://www.uvic.ca/research/centres/arc) and a member of
[CANFAR, the Canadian Advanced Network for Astronomy
Research](http://www.canfar.net).

WENDI is one of CSA's CANFAR projects. An initial version has been
developed as part of a CANFAR [Canarie](https://www.canarie.ca)
project. Its goal was to provide remote ipython notebook web-based
analysis access to the [NuGrid](http://www.nugridstars.org) stellar
nucleosynthesis data.

WENDI has been expanded in 2016 to accommodate the needs of the PPMstar
collaboration between the LCSE at the University of Minnesota and CSA
at UVic. This extension was part of a NSF [JINA-CEE](http://www.jinaweb.org/) 
project in which
both groups are participating, and supported by the NSF CDS&E program.
